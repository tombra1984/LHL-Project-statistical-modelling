# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import pandas as pd
```

```python
import requests

#get data for certain network_id
network_id = 'dublinbikes'
network_json = requests.get("http://api.citybik.es/v2/networks/dublinbikes").json()
#network_json
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
# city_bikes_df = pd.DataFrame(stations, columns = ['name', 'longitude', 'latitude', 'free_bikes'])
# city_bikes_df
```

```python
station_name = []
latitude = []
longitude = []
num_bikes = []
for station in network_json['network']['stations']:
    station_name.append( station['name'] )
    latitude.append(station['latitude'] )
    longitude.append(station['longitude'] )
    num_bikes.append(station['free_bikes'] )
```

Put your parsed results into a DataFrame.

```python
import pandas as pd
```

```python
data_city_bikes = pd.DataFrame({'name':station_name,'latitude':latitude,'longitude':longitude,'num_bikes':num_bikes})
data_city_bikes.head()
```

```python
data_city_bikes.to_csv('data_city_bikes.csv')
```
